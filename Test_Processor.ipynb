```python
import os
import re
import random
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

%matplotlib inline
```

```python
loc = 'Test_Data/'
Test = pd.DataFrame()
Files = os.listdir(loc)
```

```python
for i in range(len(Files)):

    cat_name = re.findall(r'(ZIF-\d+)\s.*.txt',Files[i])[0]

    with open(loc + Files[i],'r') as f:
        data = f.readlines()

    new_data = []
    for d in data:
        try:
            l = d.split()
            if l != []:
                new_data.append((float(l[0]), float(l[1])))
        except:
            l = d.split(',')
            if l != []:
                new_data.append((float(l[0]), float(l[1])))

    data = new_data

    X = [(lambda x: x[0])(d) for d in data]
    Y = [(lambda x: x[1])(d) for d in data]
    # Y = (Y - np.min(Y)) / (np.max(Y) - np.min(Y))
    
    df = {}
    for grid, value in zip(X, Y):
        df[grid] = [value]

    cols = [(lambda x: 'grid_' + str(x))(l) for l in X]
    Df = pd.DataFrame(df)
    Df.columns = cols

    Df['label'] = cat_name

    Test= pd.concat([Test, Df], axis = 0)

Test = Test.reset_index()
Test = Test.drop(['index'], axis = 1)
Test.head()
```

```python
with open('T_origin.pickle', 'wb') as f:
    pickle.dump(Test, f)
```
